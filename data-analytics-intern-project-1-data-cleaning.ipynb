```python
#Load Libraries for Data Analysis
import pandas as pd
import os
os.listdir('/kaggle/input/datasets/gentle419/stocksdataset/')
```

```python
data = pd.read_csv('/kaggle/input/datasets/gentle419/stocksdataset/Stock_Prices.csv')
data.head(5)
```

```python
#Check for exisiting values (True/False
data.isnull()
```

```python
#Check for exisiting values (True/False
data.isnull().sum()
```

```python
# Check how many records there are
data.shape
```

```python
# Drop rows with missing values since not much i.e 27
data_clean = data.dropna()

# Check to see output
data_clean.isnull().sum()
```

```python
# Check how many records remain
data_clean.shape
```

```python
# Check for duplicate - show true for duplicate rows
data_clean.duplicated()
```

```python
# Check for No of Duplicates
data_clean.duplicated().sum()
```

```python
# Display first 10 records
data_clean.head(10)
```

```python
# standardize date records
data_clean.loc[:, 'date'] = pd.to_datetime(data_clean['date'])
```

```python
# Display first 10 records

data_clean.head(10)
```

```python
# To Standardize Text Data
# Strip() - remove spaces
# Title() - capitalizes properly
data_clean.loc[:,  'symbol'] = data_clean['symbol'].str.strip().str.title()
```

```python
# Display first 10 records
data_clean.head(10)
```

```python
# convert numeric column
data_clean.loc[:, 'volume'] = pd.to_numeric(data_clean['volume'], errors='coerce')
```

```python
# Display first 10 records
data_clean.head(10)
```

```python
# Export Clean & Processed data
# Index = False : does not save row nos as column
data_clean. to_csv("Project1.csv", index=False)
print("Data Saved successfully")
```
