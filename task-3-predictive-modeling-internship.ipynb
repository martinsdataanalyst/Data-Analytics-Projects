```python
#Load Libraries for Data Analysis
import pandas as pd
import os
os.listdir('/kaggle/input/datasets/gentle419/sentimentdataset/')
```

```python
data = pd.read_csv("/kaggle/input/datasets/gentle419/sentimentdataset/Sentiment.csv")
print(data.head())
print(data.info())
```

```python
# Counting missing values per column
data.isnull().sum()
```

```python
# Handle Missing Values
data = data.dropna()
```

```python
#Encode Categorical Variables

# If your dataset has text labels (e.g., positive/negative):
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

# Replace 'target_column' with your actual sentiment column name
data['Sentiment'] = label_encoder.fit_transform(data['Sentiment'])

data.head()
```

```python
data = data.drop(['Unnamed: 0.1', 'Unnamed: 0'], axis=1)
data.head()
```

```python
# Separate Features and Target
X = data.drop('Sentiment', axis=1)
y = data['Sentiment']
```

```python
#Convert Other Categorical Columns (if any)
X = pd.get_dummies(X)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Train-Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
#Train Models
# Logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
#Decision Tree
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
```

```python
# Evaluation Function
def evaluate_model(y_test, y_pred, model_name):
    print(f"\n{model_name} Performance:")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Precision:", precision_score(y_test, y_pred, average='weighted'))
    print("Recall:", recall_score(y_test, y_pred, average='weighted'))
    print("F1 Score:", f1_score(y_test, y_pred, average='weighted'))
    print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
