```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
os.listdir('/kaggle/input/datasets/gentle419/stocksdataset/')
```

```python
data = pd.read_csv('/kaggle/input/datasets/gentle419/stocksdataset/Stock_Prices.csv')
data.head(5)
```

```python
# Calculate Summary Statistics
data.describe()
```

```python
# Histogram shows distribution of one column
import matplotlib.pyplot as plt
plt.hist(data['open'],bins=10, alpha=0.5, label='high', color='blue', edgecolor='black')
plt.title('Volume of Stocks Trading')
plt.show()
```

```python
# Histogram shows distribution of one column
plt.hist(data['high'], bins=10)
plt.title('Distribution of High Prices')
plt.show()
```

```python
# Boxplots compares several columns
plt.boxplot([data['open'], data['high'], data['low'], data['close']])
plt.xticks([1,2,3,4], ['open','high','low','close'])
plt.title('Price Comparision')
plt.show()
```

```python
# Scatter Plots shows relationship between Two columns
# It's used for seeing relationships i.e correlations, trends

plt.scatter(data['high'], data['low'])
plt.xlabel('high')
plt.ylabel('low')
plt.title('High vs Low Stocks Prices')
plt.show()
```

```python
# Scatter Plots shows relationship between Two columns
# It's used for seeing relationships i.e correlations, trends

plt.scatter(data['open'], data['close'])
plt.xlabel('open')
plt.ylabel('close')
plt.title('Open vs Close Stocks Prices')
plt.show()
```

Correlation Measures how strongly two variables moves together

Positive Correlation: Both increases together
Negative Correlation: One increases while the other decreases
No Correlation: No Consistent Relationship

```python
# Correlation Matrix is used to find correlations between numerical features
features = ['open', 'high', 'low', 'close', 'volume']
corr_matrix = data[features].corr()
print(corr_matrix)
```
